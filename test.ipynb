```python
import os

txt_path = r"E:\Amir\EWTracker\yolov7_for_segmentation\yolov7-segmentation\runs\predict-seg\exp15\labels\b_246.txt"

with open(txt_path, 'r') as f:
    data = f.readlines()

data = [x.strip() for x in data]

data = [x for x in data if int(x.split(' ')[2]) > 10]

data = [x for x in data if int(x.split(' ')[2]) > 100]

data = [x for x in data if int(x.split(' ')[2]) < 200]


data = [x.split(' ')[1:] for x in data]
```

```python
segment_1 = data[0]
segment_2 = data[1]
```

```python
segment_1 = [(int(segment_1[i]), int(segment_1[i+1])) for i in range(0, len(segment_1), 2)]
segment_2 = [(int(segment_2[i]), int(segment_2[i+1])) for i in range(0, len(segment_2), 2)]
```

```python
video_path = r"E:\Amir\EWTracker\yolov7_for_segmentation\yolov7-segmentation\4.mp4"

import cv2
from matplotlib import pyplot as plt

cap = cv2.VideoCapture(video_path)

# Take out the frame number 247 as image
cap.set(cv2.CAP_PROP_POS_FRAMES, 247)
ret, frame = cap.read()

plt.imshow(frame)
```

```python
lines_1 = []
lines_2 = []

for i, start_point in enumerate(segment_1):
    if i == len(segment_1) - 1:
        break
    end_point = segment_1[i+1]

    # draw line from start_point to end_point
    line = plt.Line2D((start_point[0], end_point[0]), (start_point[1], end_point[1]), lw=5., color='red')
    lines_1.append(line)

for i, start_point in enumerate(segment_2):
    if i == len(segment_2) - 1:
        break
    end_point = segment_2[i+1]

    # draw line from start_point to end_point
    line = plt.Line2D((start_point[0], end_point[0]), (start_point[1], end_point[1]), lw=5., color='blue')
    lines_2.append(line)

# set fig size = 20, 20
fig, ax = plt.subplots(figsize=(20, 20))


plt.imshow(frame)

for line in lines_1:
    ax.add_line(line)

plt.show()
```

```python
#Given two integer numbers return their product only if the product is equal to or lower than 1000, else return their sum.
#TAKE input of two numbers from user

def product_or_sum(num1, num2):
    if num1 * num2 <= 1000:
        return num1 * num2
    else:
        return num1 + num2
#TAKE input of two numbers from user and call the function
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
result = product_or_sum(num1, num2)
print("The result is", result)


```
